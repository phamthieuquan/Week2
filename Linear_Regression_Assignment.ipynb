## Training and Testing Data

**8.** Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.
** Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column. **
5pts

```python
# New technic: train_test_split splits arrays or matrices into random train and test subsets
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(d[['Time on App']], d['Yearly Amount Spent'], test_size=0.3, random_state=101)
# Notice that we use [['X']] and ['Y'] here. Any idea why?

# We sort the training set in order to keep the curve-form when we plot it on a graph using plt.plot()
X_train = X_train.sort_index()
y_train = y_train.sort_index()
```

```python

plt.figure(figsize=(20, 10))

# Your code here
plt.scatter(X_train,y_train,label='train')
plt.scatter(X_test,y_test,label='test')
plt.legend()
plt.show()
```
